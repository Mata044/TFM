```python
def candidates(n):
    L=[]
    for r in prime_range(n):
        for s in prime_range(r):
            if (r+s).prime_to_m_part(2).is_prime() and (r-s).prime_to_m_part(2).is_prime():
                L.append((r,s))
    return L
```
